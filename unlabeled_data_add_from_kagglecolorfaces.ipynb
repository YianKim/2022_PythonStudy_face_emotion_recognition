데이터로 바꿔주기

face1 ~ face67224

```python
import numpy as np
```

```python
face_array=[]

for n in range(1,67225):
  img_face = Image.open('/content/drive/MyDrive/파이썬스터디 프로젝트/Humanresizes/face' + str(n) + '.jpg')
  face_array.append(np.array(img_face))
```

```python
face_array = np.array(face_array)
```

```python
face_array = face_array/255
```

```python
face_array2 = np.reshape(face_array, (67224, 48, 48, 1))
```

```python
face_array2.shape
```

```python
import pickle

with open('/content/drive/MyDrive/파이썬스터디 프로젝트/face_array.pkl', 'wb') as f:
	pickle.dump(face_array2, f, protocol=pickle.HIGHEST_PROTOCOL)
```
